# 1) getting the saved list cars

```python
import pickle
with open("test", "rb") as data:   
      cars= pickle.load(data)
```

# 2) number of the cars in the list

```python
len(cars)
```

# 3) creating separate lists for each information of the car

```python
make=[]
model=[]
usedsince=[]
km=[]
trans=[]
city=[]
color=[]
fuel=[]
comfort=[]
window=[]
sound=[]
other=[]
price=[]
for car in cars:
    res=any("Make" in string for string in car)
    if res == True:
        make.append( list(filter(lambda x: 'Make' in x, car))[0])
    else:
        make.append('Nan')
    
    res=any("Model" in string for string in car)
    if res == True:
        model.append( list(filter(lambda x: 'Model' in x, car))[0])
    else:
        model.append('Nan')
    
    res=any("Used since" in string for string in car)
    if res == True:
        usedsince.append( list(filter(lambda x: 'Used since' in x, car))[0])
    else:
        usedsince.append('Nan')
    
    res=any("Km" in string for string in car)
    if res == True:
        km.append( list(filter(lambda x: 'Km' in x, car))[0])
    else:
        km.append('Nan')
        
    res=any("Transmission" in string for string in car)
    if res == True:
        trans.append( list(filter(lambda x: 'Transmission' in x, car))[0])
    else:
        trans.append('Nan')
        
        
        
    res=any("City" in string for string in car)
    if res == True:
        city.append( list(filter(lambda x: 'City' in x, car))[0])
    else:
        city.append('Nan')
        
    res=any("Color" in string for string in car)
    if res == True:
        color.append( list(filter(lambda x: 'Color' in x, car))[0])
    else:
        color.append('Nan')
        
    res=any("Fuel" in string for string in car)
    if res == True:
        fuel.append( list(filter(lambda x: 'Fuel' in x, car))[0])
    else:
        fuel.append('Nan')
        
    res=any("Comfort" in string for string in car)
    if res == True:
        comfort.append( list(filter(lambda x: 'Comfort' in x, car))[0])
    else:
        comfort.append('Nan')
        
    res=any("Windows" in string for string in car)
    if res == True:
        window.append( list(filter(lambda x: 'Windows' in x, car))[0])
    else:
        window.append('Nan')
        
    res=any("Sound System" in string for string in car)
    if res == True:
        sound.append( list(filter(lambda x: 'Sound System' in x, car))[0])
    else:
        sound.append('Nan')
        
    res=any("Other" in string for string in car)
    if res == True:
        other.append( list(filter(lambda x: 'Other' in x, car))[0])
    else:
        other.append('Nan')
    
    if 'EGP' in car[-1]:
        price.append(car[-1])
    else:
        price.append('Nan')
    
    

        
    
       
```

# 4) creating data frame 

```python
import pandas as pd
data = {'Make': make,'Model': model,'UsedSince': usedsince,'KiloMeters':km,
        'Transmission':trans,'City':city,'Color':color,'Fuel':fuel,
        'Comfort':comfort,'Windows':window,'Sound':sound,'Others':other,
        'Price':price}
        
   
 df= pd.DataFrame(data)
```
